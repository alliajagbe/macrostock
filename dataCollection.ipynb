```python
import pandas as pd
import yfinance as yf
```

```python
start_date = "2000-01-01"
end_date = "2024-11-30"
```

```python
def fetch_stock_data(ticker, start_date, end_date):
    data = yf.download(ticker, start=start_date, end=end_date)
    daily_returns = data['Adj Close'].pct_change()
    data['Daily_Returns'] = daily_returns
    return data
```

```python
# s&p 500
sp500 = fetch_stock_data('^GSPC', start_date, end_date)
sp500.head()
```

```python
# making Date as index
cpi.set_index('DATE', inplace=True)
consumer_sentiment.set_index('DATE', inplace=True)
dexuseu.set_index('DATE', inplace=True)
effr.set_index('DATE', inplace=True)
gdp.set_index('DATE', inplace=True)
unrate.set_index('DATE', inplace=True)
sp500.set_index(sp500.index, inplace=True)
usdx.set_index(usdx.index, inplace=True)
vix.set_index(vix.index, inplace=True)
```

```python
# parsing date
cpi.index = pd.to_datetime(cpi.index)
consumer_sentiment.index = pd.to_datetime(consumer_sentiment.index)
dexuseu.index = pd.to_datetime(dexuseu.index)
effr.index = pd.to_datetime(effr.index)
gdp.index = pd.to_datetime(gdp.index)
unrate.index = pd.to_datetime(unrate.index)
sp500.index = pd.to_datetime(sp500.index)
usdx.index = pd.to_datetime(usdx.index)
vix.index = pd.to_datetime(vix.index)
```

```python
cpi_daily = cpi.resample('D').ffill()
consumer_sentiment_daily = consumer_sentiment.resample('D').ffill()
unrate_daily = unrate.resample('D').ffill()
gdp_daily = gdp.resample('D').ffill()
```

```python
datasets = [sp500, vix, dexuseu, effr, usdx, cpi_daily, consumer_sentiment_daily, unrate_daily, gdp_daily]
final_dataset = pd.concat(datasets, axis=1)
```

```python
columns = [
    ('Adj Close', '^GSPC'), ('Close', '^GSPC'), ('High', '^GSPC'),
    ('Low', '^GSPC'), ('Open', '^GSPC'), ('Volume', '^GSPC'),
    ('Daily_Returns', ''), ('Adj Close', '^VIX'), ('Close', '^VIX'),
    ('High', '^VIX'), ('Low', '^VIX'), ('Open', '^VIX'),
    ('Volume', '^VIX'), ('Daily_Returns', ''), 'DEXUSEU',
    'EFFR', 'DX-Y.NYB', 'CPALTT01USM657N', 'CSCICP03USM665S', 'UNRATE', 'GDP'
]

# Rename the columns to more meaningful names
renamed_columns = {
    ('Adj Close', '^GSPC'): 'S&P500_Adj_Close',
    ('Close', '^GSPC'): 'S&P500_Close',
    ('High', '^GSPC'): 'S&P500_High',
    ('Low', '^GSPC'): 'S&P500_Low',
    ('Open', '^GSPC'): 'S&P500_Open',
    ('Volume', '^GSPC'): 'S&P500_Volume',
    ('Daily_Returns', ''): 'S&P500_Daily_Returns',
    ('Adj Close', '^VIX'): 'VIX_Adj_Close',
    ('Close', '^VIX'): 'VIX_Close',
    ('High', '^VIX'): 'VIX_High',
    ('Low', '^VIX'): 'VIX_Low',
    ('Open', '^VIX'): 'VIX_Open',
    ('Volume', '^VIX'): 'VIX_Volume',
    ('Daily_Returns', ''): 'VIX_Daily_Returns',
    'DEXUSEU': 'USD_EUR_Exchange_Rate',
    'EFFR': 'Effective_Federal_Funds_Rate',
    'DX-Y.NYB': 'US_Dollar_Index',
    'CPALTT01USM657N': 'CPI_All_Items',
    'CSCICP03USM665S': 'Consumer_Sentiment_Index',
    'UNRATE': 'Unemployment_Rate',
    'GDP': 'GDP'
}
```

```python
final_dataset.columns = [renamed_columns[col] if col in renamed_columns else col for col in final_dataset.columns]
```
