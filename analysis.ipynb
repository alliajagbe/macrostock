```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller, grangercausalitytests
from arch import arch_model
from statsmodels.formula.api import ols
```

```python
# parse dates
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
# Generate returns
df['S&P500_Daily_Returns'] = df['S&P500_Adj_Close'].pct_change().dropna()
df['VIX_Daily_Returns'] = df['VIX_Adj_Close'].pct_change().dropna()
```

```python
# log gdp 
df['Log_GDP'] = np.log(df['GDP'])
```

```python
# -------------------------------
# Exploratory Data Analysis (EDA)
# -------------------------------

# Plot trends
plt.figure(figsize=(10, 6))
df['S&P500_Adj_Close'].plot(title="S&P500")
plt.show()

plt.figure(figsize=(10, 6))
df['VIX_Adj_Close'].plot(title="VIX")
plt.show()

# Correlation heatmap
plt.figure(figsize=(20, 10))
columns_to_take = ['S&P500_Adj_Close', 'VIX_Adj_Close', 'USD_EUR_Exchange_Rate', 
                   'Effective_Federal_Funds_Rate', 'S&P500_Volume', 'US_Dollar_Index', 
                   'CPI_All_Items', 'Consumer_Sentiment_Index', 'Unemployment_Rate', 
                   'Log_GDP', 'VIX_Daily_Returns', 'S&P500_Daily_Returns']
correlation_matrix = df[columns_to_take].corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# -------------------------------
# Volatility Modeling (ARCH/GARCH)
# -------------------------------
# Fit a GARCH model to S&P500 returns
garch_model = arch_model(df['S&P500_Daily_Returns'].dropna(), vol='Garch', p=1, q=1)
garch_fit = garch_model.fit()
print(garch_fit.summary())

# Plot volatility
plt.figure(figsize=(10, 6))
plt.plot(garch_fit.conditional_volatility, label="Conditional Volatility")
plt.title("GARCH Model - Conditional Volatility")
plt.legend()
plt.show()
```

```python
# -------------------------------
# Multivariate Time Series (VAR)
# -------------------------------
# Select relevant columns for VAR
macro_vars = ['CPI_All_Items', 'Consumer_Sentiment_Index', 'Unemployment_Rate', 'GDP', 'US_Dollar_Index', 'USD_EUR_Exchange_Rate', 'Effective_Federal_Funds_Rate']
selected_data = df[['S&P500_Adj_Close'] + macro_vars].dropna()

selected_data_log = selected_data.applymap(lambda x: np.log(x) if x > 0 else np.log(1e-6))

# Fit VAR model
var_model = VAR(selected_data_log)
var_results = var_model.fit(maxlags=5)
print(var_results.summary())
```

```python
# Impulse response
irf = var_results.irf(10)
irf.plot(orth=True, figsize=(40, 40))
plt.tight_layout()
plt.show()
```

```python
# -------------------------------
# Granger Causality
# -------------------------------
# Granger causality between CPI and S&P500
granger_results = grangercausalitytests(selected_data[['CPI_All_Items', 'S&P500_Adj_Close']], maxlag=5)
```

```python
# -------------------------------
# Cointegration Analysis
# -------------------------------
from statsmodels.tsa.vector_ar.vecm import coint_johansen

coint_result = coint_johansen(selected_data[['S&P500_Adj_Close', 'GDP']], det_order=0, k_ar_diff=1)
print(f"Trace statistic: {coint_result.lr1}")
print(f"Critical values: {coint_result.cvt}")
```

```python
# -------------------------------
# Portfolio Analysis
# -------------------------------
# Calculate portfolio metrics
mean_return = df['S&P500_Daily_Returns'].mean()
std_dev = df['S&P500_Daily_Returns'].std()
sharpe_ratio = mean_return / std_dev

print(f"Mean Return: {mean_return:.4f}, Volatility: {std_dev:.4f}, Sharpe Ratio: {sharpe_ratio:.2f}")
```

```python
# -------------------------------
# Value at Risk (VaR)
# -------------------------------
# Historical VaR at 95% confidence
var_95 = np.percentile(df['S&P500_Daily_Returns'].dropna(), 5)
print(f"Historical VaR (95%): {var_95:.4f}")

# Parametric VaR
parametric_var = mean_return - 1.65 * std_dev
print(f"Parametric VaR (95%): {parametric_var:.4f}")
```
